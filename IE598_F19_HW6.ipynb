```python
import pandas as pd
df = pd.read_csv('C:/Users/11816/Desktop/mechine learning/HW6/ccdefault.csv')
df.info()
```

```python
da=df.iloc[:,1:]
da.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
X = da.iloc[:, :-1].values
y = da['DEFAULT'].values
train_score=[]
test_score=[]

for i in range(1,11):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,stratify=y,random_state=i)
    tree = DecisionTreeClassifier(criterion='gini',max_depth=4,random_state=i)
    tree.fit(X_train,y_train)
    train_score.append(tree.score(X_train,y_train))
    test_score.append(tree.score(X_test,y_test))

for i in range(1,11):
    print('train accuracy score for random_state= %.i is: %.6f' % (i,train_score[i-1]))
print('mean score: %.6f' % np.mean(train_score))
print('standard deviation: %.6f' % np.std(train_score))

for i in range(1,11):
    print('test accuracy score for random_state= %.i is: %.6f' % (i,test_score[i-1]))
print('mean score: %.6f' % np.mean(test_score))
print('standard deviation: %.6f' % np.std(test_score))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

train_cv_scores=[]
for i in range(1,11):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,stratify=y,random_state=i)
    kfold=StratifiedKFold(n_splits=10,random_state=i).split(X_train,y_train)
    scores=cross_val_score(estimator=tree,X=X_train,y=y_train,cv=10,n_jobs=1)
    train_cv_scores.append(scores)
    print('when random_state=%.i :' % i)
    print('cv accuracy scores for train data: %s' % train_cv_scores[i-1])
    print('mean cv score for train data: %.6f' % np.mean(train_cv_scores[i-1]))
    print('standard deviation for cv score for train data: %.6f' % np.std(train_cv_scores[i-1]))
    print(' ')
```
